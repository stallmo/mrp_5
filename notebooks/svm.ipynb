```python
import seaborn as sns
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, "../code/")
from svm_ann import main
import RandomForestClassifier as rfc
import GpClassifier as gpc
```

```python
seed = 42
big5 = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness to experience']
tasks = ['find keys', 'find hidden object', 'look for non-existing object', 'remember drawer content', 'write', 'make tea']
titlesize=15
labelsize=10
```

```python
predictions = main("../pickle_data/feature_dataframes/all_features_restructured.p", seed, False)
plt.figure(figsize=(15,10))
sns.heatmap(predictions, annot=True, xticklabels=big5, yticklabels=tasks)
plt.title("SVM accuracies per trait per task", fontsize=titlesize)
plt.xlabel('Trait', fontsize=labelsize)
plt.ylabel('Task', fontsize=labelsize)
plt.show()
```

### Random Forest

```python
predictions = rfc.main("../pickle_data/feature_dataframes/all_features_restructured.p", seed, False)
plt.figure(figsize=(15,10))
sns.heatmap(predictions, annot=True, xticklabels=big5, yticklabels=tasks)
plt.title("RandomForest accuracies per trait per task", fontsize=titlesize)
plt.xlabel('Trait', fontsize=labelsize)
plt.ylabel('Task', fontsize=labelsize)
plt.show()
```
